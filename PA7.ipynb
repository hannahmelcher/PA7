```python
import math 

class calculator():
    def __init__(self, input1, input2):
        self.input1=input1
        self.input2=input2
    def addition(self):
        return self.input1+self.input2
    def subtraction(self):
        return self.input1-self.input2
    def multiplication(self):
        return self.input1*self.input2
    def division(self):
        return self.input1/self.input2
    def mean(self):
        return (self.input1+self.input2)/2
    def exponent(self):
        return self.input1 ** self.input2
    def square_root(self):
        return math.sqrt(self.input1)
    def sin(self):
        return math.sin(self.input1)
    def cos(self):
        return math.cos(self.input1)
    def tan(self):
        return math.tan(self.input1)
    def log(self):
        if self.input1 <= 0:
            return "Invalid input"
        return math.log10(self.input1)
    def natural_log(self):
        if self.input1 <= 0:
            return "Invalid input"
        return math.log(self.input1)
    def ceiling(self):
        return math.ceil(self.input1)
    def hyperbolic_cosine(self):
        return math.cosh(self.input1)
    def factorial(self):
        return math.factorial(self.input1)
    
input_1=int(input("Enter first number: "))
input_2=int(input("Enter second number: "))
my_calc=calculator(input_1,input_2)

choice=1
while choice!=0:
    print()
    print("0. Exit")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Mean")
    print("6. Exponent")
    print("7. Square Root")
    print("8. Sine")
    print("9. Cosine")
    print("10. Tangent")
    print("11. Log")
    print("12. Natural Log")
    print("13. Ceiling")
    print("14. Hyberbolic Cosine")
    print("15. Factorial")
    choice=int(input("Enter your choice: "))
    
    if choice==1:
        print(f"Result: {my_calc.addition()}")
        
    elif choice==2:
        print(f"Result: {my_calc.subtraction()}")
        
    elif choice==3:
        print(f"Result: {my_calc.multiplication()}")
        
    elif choice==4:
        print(f"Result: {my_calc.division()}")
        
    elif choice==5:
        print(f"Result: {my_calc.mean()}")
        
    elif choice==6:
        print(f"Result: {my_calc.exponent()}")
        
    elif choice==7:
        print(f"Result: {my_calc.square_root()}")
        
    elif choice==8:
        print(f"Result: {my_calc.sin()}")
        
    elif choice==9:
        print(f"Result: {my_calc.cos()}")
        
    elif choice==10:
        print(f"Result: {my_calc.tan()}")
        
    elif choice==11:
        print(f"Result: {my_calc.log()}")
        
    elif choice==12:
        print(f"Result: {my_calc.natural_log()}")
        
    elif choice==13:
        print(f"Result: {my_calc.ceiling()}")
        
    elif choice==14:
        print(f"Result: {my_calc.hyperbolic_cosine()}")
        
    elif choice==15:
        print(f"Result: {my_calc.factorial()}")
        
    elif choice==0:
        print("Exit. Thank you! :)")
        
    else:
        print("Sorry. Invalid choice.")
        
print()
```
